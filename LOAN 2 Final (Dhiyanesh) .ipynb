#DATA CLEANING

```python
# Check for the missing or the null values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check for any duplicate values
duplicates = df.duplicated(subset='Customer ID').sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# If present drop the duplicate values
# df = df.drop_duplicates(subset='Customer ID')
```

```python
# Fill missing values for categorical columns
df['Gender'] = df['Gender'].fillna('Rather not say')
df['Type of Employment'] = df['Type of Employment'].fillna('Unknown')
df['Has Active Credit Card'] = df['Has Active Credit Card'].fillna('Unknown')
df['Income Stability'] = df['Income Stability'].fillna('Unknown')
df['Property Location'] = df['Property Location'].fillna('Unknown')
```

```python
# Fill missing values for numerical columns with the median
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Income (USD)'] = df['Income (USD)'].fillna(df['Income (USD)'].median())
df['Loan Amount Request (USD)'] = df['Loan Amount Request (USD)'].fillna(df['Loan Amount Request (USD)'].median())
df['Current Loan Expenses (USD)'] = df['Current Loan Expenses (USD)'].fillna(df['Current Loan Expenses (USD)'].median())
df['Credit Score'] = df['Credit Score'].fillna(df['Credit Score'].median())
df['Property Age'] = df['Property Age'].fillna(df['Property Age'].median())
df['Property Price'] = df['Property Price'].fillna(df['Property Price'].median())
df['Loan Sanction Amount (USD)'] = df['Loan Sanction Amount (USD)'].fillna(df['Loan Sanction Amount (USD)'].median())
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].median())
```

```python
df.isnull().sum()
```

```python
# Ensure correct data types
df['Age'] = df['Age'].astype(int)
df['Income (USD)'] = df['Income (USD)'].astype(float)
df['Loan Amount Request (USD)'] = df['Loan Amount Request (USD)'].astype(float)
df['Current Loan Expenses (USD)'] = df['Current Loan Expenses (USD)'].astype(float)
df['Credit Score'] = df['Credit Score'].astype(float)
df['Property Age'] = df['Property Age'].astype(float)
df['Property Price'] = df['Property Price'].astype(float)
df['Loan Sanction Amount (USD)'] = df['Loan Sanction Amount (USD)'].astype(float)
```

#OUTLIER DETECTION AND REMOVAL

```python
# before outliers removal(outliers present)
sns.boxplot(df["Income (USD)"])
```

```python
#Finding the no of outliers
df = pd.DataFrame(df)
first_quartile = df["Income (USD)"].quantile(0.25)
third_quartile = df["Income (USD)"].quantile(0.75)
interquartile_range = third_quartile - first_quartile
upper_bound = float(third_quartile + 1.5 * interquartile_range)
lower_bound = float(first_quartile - 1.5 * interquartile_range)
outliers = df[(df["Income (USD)"] < lower_bound) | (df["Income (USD)"] > upper_bound)]["Income (USD)"]
print("Number of outliers:", outliers.shape[0])
```

```python
#Removing the outliers 

df["Income (USD)"] = df["Income (USD)"].clip(lower=lower_bound, upper=upper_bound)
sns.boxplot(df["Income (USD)"])
```

```python
#Loan Amount Request (USD) is a numeric value so boxplot is possible to check outliers
sns.boxplot(df["Loan Amount Request (USD)"])
```

#UNIVARIATE ANALYSIS

Univariate analysis focuses on analyzing each variable individually. It provides insights into the distribution, central tendency, and variability of each variable. For the given dataset, we'll perform univariate analysis on both numerical and categorical features.

```python
# Summary statistics for numerical variables
numerical_summary = df.describe()

# Summary statistics for categorical variables
categorical_summary = df.describe(include=['object'])

numerical_summary, categorical_summary
```

```python
# List of numerical columns to plot histograms for
numerical_columns = ['Age', 'Income (USD)', 'Loan Amount Request (USD)', 'Credit Score', 'Property Price', 'Loan Sanction Amount (USD)']

# Plot histograms
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[column].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# List of categorical columns to plot bar plots for
categorical_columns = ['Gender', 'Income Stability', 'Profession', 'Type of Employment', 'Location']

# Plot bar plots
plt.figure(figsize=(15, 10))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=column, data=df, order=df[column].value_counts().index)
    plt.title(f'Count of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
